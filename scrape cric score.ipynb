```python
#DEFINE function to prepare the batting and bowling data

def scorebatting(innings):
    
    table = soup.find_all('table')

    all_tbody = table[innings].find_all('tbody')

    score = []

    for tbody in all_tbody:
        for tr in tbody:
            pv = [item.text.strip() for item in tr]
            if len(pv) > 1:
                score.append(pv)


    batsmen_df = pd.DataFrame(columns=["Name","Desc","Runs", "Balls", "M",  "4s", "6s", "SR"])


    for i in score:
        if len(i) == 8:
            df = pd.Series(i,index=batsmen_df.columns)
            batsmen_df = batsmen_df.append(df, ignore_index=True)
            
    return batsmen_df


def scorebowling(innings):
    
    table = soup.find_all('table')

    all_tbody = table[innings].find_all('tbody')

    score = []

    for tbody in all_tbody:
        for tr in tbody:
            pv = [item.text.strip() for item in tr]
            if len(pv) > 1:
                score.append(pv)


    bowler_df = pd.DataFrame(columns=['Name', 'O', 'M', 'R', 'W','Econ', 'WD', 'NB'])


    for i in score:
        if len(i) == 11:
            x = [i[0],i[1],i[2],i[3],i[4],i[5],i[9],i[10]]
            df = pd.Series(x,index=bowler_df.columns)
            bowler_df = bowler_df.append(df, ignore_index=True)
            
    return bowler_df
```
